```python
import os
import time
import random
from collections import deque
from concurrent.futures import ThreadPoolExecutor

import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import matplotlib.pyplot as plt

import google.generativeai as genai
import PIL.Image

from google.api_core.exceptions import (
    TooManyRequests,
    ServiceUnavailable,
    InternalServerError,
    Aborted,
)

# --- Configuration ---
# IMPORTANT: Set your GOOGLE_API_KEY as an environment variable
# or replace the default value below.
MY_API_KEY         = os.getenv("GOOGLE_API_KEY", "AIzaSyDjMJ_1fymbbG2B51gZ4zDyVqQbmzZ1S1E")
MY_GEMINI_MODEL    = "gemini-1.5-flash-latest"
RATE_LIMIT_RPM     = 30                # Max calls per 60s window for the Gemini API

# --- Experiment Parameters ---
initial_training_size = 500             # Start with a smaller initial set to see the curves grow
total_queries        = 15               # More queries to see a longer learning curve
query_batch_size     = 100              # Number of items to query in each loop
num_classes          = 10
input_shape          = (28, 28, 1)


# --- Configure the Gemini client ---
genai.configure(
    api_key=MY_API_KEY,
    client_options={"api_endpoint": "https://generativelanguage.googleapis.com"},
)


# --- MNIST Data Preparation ---
(x_full, y_full), (x_test, y_test) = keras.datasets.mnist.load_data()
# Normalize and reshape data
x_full = np.expand_dims(x_full.astype("float32") / 255.0, -1)
x_test = np.expand_dims(x_test.astype("float32") / 255.0, -1)
# Create one-hot encoded test labels
y_test_cat = keras.utils.to_categorical(y_test, num_classes)


def create_model():
    """Defines and compiles the CNN architecture."""
    model = keras.Sequential([
        keras.Input(shape=input_shape),
        layers.Conv2D(32, 3, activation="relu"),
        layers.MaxPooling2D(2),
        layers.Conv2D(64, 3, activation="relu"),
        layers.MaxPooling2D(2),
        layers.Flatten(),
        layers.Dropout(0.5),
        layers.Dense(num_classes, activation="softmax"),
    ])
    model.compile(optimizer="adam",
                  loss="categorical_crossentropy",
                  metrics=["accuracy"])
    return model


# --- Gemini API Interaction with Rate Limiting and Parallelism ---

_request_times = deque()

def _wait_for_rate_slot():
    """Ensures no more than RATE_LIMIT_RPM calls in any rolling 60s window."""
    now = time.time()
    _request_times.append(now)
    while _request_times and now - _request_times[0] > 60:
        _request_times.popleft()
    if len(_request_times) > RATE_LIMIT_RPM:
        earliest = _request_times[0]
        to_sleep = 60 - (now - earliest) + 0.01
        print(f"Rate limit hit: sleeping {to_sleep:.2f}s")
        time.sleep(to_sleep)


def _label_single_image_with_retry(model, img_array, prompt_text):
    """Calls Gemini with exponential backoff and our rate limiter."""
    pil_img = PIL.Image.fromarray((np.squeeze(img_array) * 255).astype(np.uint8))
    max_retries = 5
    base_wait = 2.0
    for attempt in range(max_retries):
        _wait_for_rate_slot()
        try:
            response = model.generate_content([prompt_text, pil_img])
            txt = response.text.strip()
            # Default to 0 if parsing fails, with a warning
            if not txt.isdigit():
                print(f"  → Warning: Could not parse '{txt}'. Defaulting to 0.")
                return 0
            return int(txt)
        except (TooManyRequests, ServiceUnavailable, InternalServerError, Aborted) as e:
            wait = base_wait ** (attempt + 1) + random.random()
            print(f"  → API {type(e).__name__}; backing off {wait:.1f}s")
            time.sleep(wait)
        except Exception as e:
            print(f"  → An unexpected error occurred: {e}. Defaulting to 0.")
            return 0
    # All retries failed
    return 0


def label_with_llm_gemini_parallel(x_batch):
    """Labels a batch by submitting images to Gemini in parallel using a ThreadPoolExecutor."""
    model = genai.GenerativeModel(MY_GEMINI_MODEL)
    prompt_text = "What digit 0–9 is in this image? Reply with just the single digit."
    labels = []
    with ThreadPoolExecutor(max_workers=RATE_LIMIT_RPM) as executor:
        # Create a future for each image labeling task
        futures = [executor.submit(_label_single_image_with_retry, model, img, prompt_text) for img in x_batch]
        # Collect results as they complete
        for future in futures:
            labels.append(future.result())
    return np.array(labels)


def compute_entropy(probs):
    """Calculates prediction entropy for a batch of probabilities."""
    return -np.sum(probs * np.log(probs + 1e-12), axis=1)


def run_experiment(mode):
    """
    Runs a full learning experiment.

    Args:
        mode (str): One of 'passive', 'active_perfect', or 'active_llm'.
    """
    if mode not in ['passive', 'active_perfect', 'active_llm']:
        raise ValueError("Mode must be 'passive', 'active_perfect', or 'active_llm'")

    # --- Initialization ---
    np.random.seed(42) # Ensure experiments are reproducible
    all_indices = np.random.permutation(len(x_full))

    train_indices = all_indices[:initial_training_size]
    pool_indices = all_indices[initial_training_size:]

    x_train = x_full[train_indices]
    y_train = y_full[train_indices]

    model = create_model()
    accuracies = []

    # --- Main Learning Loop ---
    for q in range(total_queries + 1):
        print(f"\n--- Mode: {mode} | Iteration: {q} | Training size: {len(x_train)} ---")

        # Fine-tune the model on the current training data
        y_train_cat = keras.utils.to_categorical(y_train, num_classes)
        model.fit(x_train, y_train_cat,
                  batch_size=128, epochs=10,
                  validation_split=0.1, verbose=0)

        # Evaluate and store accuracy
        loss, acc = model.evaluate(x_test, y_test_cat, verbose=0)
        accuracies.append(acc)
        print(f"Test accuracy: {acc:.4f}")

        if q == total_queries:
            break

        # --- Select and Label New Data ---
        x_pool = x_full[pool_indices]
        y_pool = y_full[pool_indices]

        if mode == 'passive':
            print(f"Selecting {query_batch_size} random samples from the pool...")
            query_indices_in_pool = np.random.choice(len(pool_indices), query_batch_size, replace=False)
        else: # Active learning modes
            print("Calculating uncertainty over the pool...")
            probs = model.predict(x_pool, batch_size=512, verbose=0)
            entropies = compute_entropy(probs)
            query_indices_in_pool = np.argsort(-entropies)[:query_batch_size]

            # **Visualization**: Show the most uncertain images for the first query
            if q == 0:
                print("Displaying 10 most uncertain images selected for labeling:")
                plt.figure(figsize=(10, 2))
                for i in range(10):
                    ax = plt.subplot(1, 10, i + 1)
                    plt.imshow(x_pool[query_indices_in_pool[i]].squeeze(), cmap="gray")
                    plt.axis("off")
                plt.suptitle(f'Uncertain Images for "{mode}" Strategy')
                plt.show()

        # Get the selected data
        x_query = x_pool[query_indices_in_pool]

        # Get labels for the selected data based on the mode
        if mode == 'active_llm':
            print(f"Querying LLM oracle for {len(x_query)} labels (in parallel)...")
            y_new = label_with_llm_gemini_parallel(x_query)
            print("LLM labeling complete.")
        else: # For passive and active_perfect, use the true labels
            print("Querying perfect oracle (ground truth labels)...")
            y_new = y_pool[query_indices_in_pool]

        # Add newly labeled data to the training set
        x_train = np.concatenate([x_train, x_query], axis=0)
        y_train = np.concatenate([y_train, y_new], axis=0)

        # Remove the queried indices from the pool
        pool_indices = np.delete(pool_indices, query_indices_in_pool, axis=0)

    return accuracies


if __name__ == "__main__":
    if MY_API_KEY == "YOUR_API_KEY_HERE":
        print("!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!")
        print("!!! WARNING: Please replace 'YOUR_API_KEY_HERE' with  !!!")
        print("!!! your actual Google API key to run the LLM oracle. !!!")
        print("!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!\n")

    # --- Run all three experiments ---
    passive_acc = run_experiment('passive')
    al_perfect_acc = run_experiment('active_perfect')
    al_llm_acc = run_experiment('active_llm')

    # --- Generate the comparative plot ---
    sizes = [initial_training_size + i * query_batch_size for i in range(total_queries + 1)]

    plt.style.use('seaborn-v0_8-whitegrid')
    plt.figure(figsize=(12, 8))

    plt.plot(sizes, passive_acc, marker='s', linestyle='--', label="Passive Learning (Random Sampling)")
    plt.plot(sizes, al_llm_acc, marker='x', linestyle='-', label="Active Learning w/ LLM Oracle")
    plt.plot(sizes, al_perfect_acc, marker='o', linestyle='-', label="Active Learning w/ Perfect Oracle (Ideal)")

    plt.xlabel("Training Set Size", fontsize=14)
    plt.ylabel("Test Accuracy", fontsize=14)
    plt.title("Active Learning Effectiveness on MNIST", fontsize=16)
    plt.legend(fontsize=12)
    plt.xticks(sizes)
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

    # --- Final Efficiency Summary ---
    target_accuracy = 0.96 # A reasonable target for this model
    print("\n" + "="*50)
    print("       Labeling Efficiency Summary")
    print("="*50)
    print(f"Labels required to reach {target_accuracy*100:.0f}% accuracy:")

    try:
        passive_size = sizes[next(i for i, acc in enumerate(passive_acc) if acc >= target_accuracy)]
        print(f"  - Passive Learning:  {passive_size} labels")
    except StopIteration:
        print(f"  - Passive Learning did not reach the target accuracy.")

    try:
        llm_size = sizes[next(i for i, acc in enumerate(al_llm_acc) if acc >= target_accuracy)]
        print(f"  - Active (LLM):      {llm_size} labels")
        if 'passive_size' in locals():
            savings = 100 * (1 - llm_size / passive_size)
            print(f"    -> A labeling cost saving of {savings:.1f}% over passive.")
    except StopIteration:
        print(f"  - Active (LLM) did not reach the target accuracy.")

    try:
        perfect_size = sizes[next(i for i, acc in enumerate(al_perfect_acc) if acc >= target_accuracy)]
        print(f"  - Active (Perfect):  {perfect_size} labels")
    except StopIteration:
        print(f"  - Active (Perfect) did not reach the target accuracy.")
    print("="*50)
```
